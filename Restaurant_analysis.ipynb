**SQL Queries**

```python
import sqlite3
import pandas as pd

# Re-defining df here to ensure it's available if preceding cells were not executed.
# In a typical notebook workflow, you would ensure the cells defining df are run first.
df = pd.read_csv('/content/Dataset .csv')
df = df.dropna()
df = df.drop_duplicates()

conn = sqlite3.connect(':memory:')
df.to_sql('restaurants', conn, index=False, if_exists='replace')
```

TOP CUISINES

```python
pd.read_sql_query("""SELECT Cuisines, COUNT(*) AS count FROM restaurants GROUP BY Cuisines ORDER BY count DESC LIMIT 3;""", conn)
```

AVERAGE RATING PER CITY

```python
pd.read_sql_query("""SELECT city, AVG("Aggregate rating") as avg_rating FROM restaurants GROUP BY city ORDER BY avg_rating DESC;""", conn)
```

ONLINE DELIVERY PERECENTAGE

```python
pd.read_sql_query("""SELECT "Has Online delivery", COUNT(*) as count, (COUNT(*) * 100.0 / (SELECT COUNT(*) FROM restaurants)) AS percentage FROM restaurants GROUP BY "Has Online delivery";""", conn)
```
